```python
from pathlib import Path
from diffusers import StableDiffusionPipeline
import torch

#Find project root
project_root = Path.cwd().resolve()

# If running from notebooks/, go one level up to project root
if project_root.name == "notebooks":
    project_root = project_root.parent

#Create /results directory
results_dir = project_root / "results"
results_dir.mkdir(parents=True, exist_ok=True)

#Load Stable Diffusion model 
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16,
).to("cuda")

#Generate image
prompt = "a speed bump on a suburban asphalt road, daylight, realistic photo"
image = pipe(prompt).images[0]

#Save image into results folder
save_path = results_dir / "baseline_speedbump.png"
image.save(save_path)
#print(f"Image saved at: {save_path}")
```
